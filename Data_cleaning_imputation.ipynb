```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coil= pd.read_csv("coil.csv")
```

```python
coil
```

```python
coil['CreationTimestamp'].min()
```

```python
coil.dtypes
```

```python
coil.isnull().sum()
```

```python
coil=coil.drop(['DisableScanning'], axis=1)
coil
```

```python
coil['CreationTimestamp'] = pd.to_datetime(coil['CreationTimestamp'])
```

```python
coil.dtypes
```

```python
coil['CreationTimestamp'].min()
```

```python
cutoff_date = pd.Timestamp('2022-08-31')
df_filtered = coil[coil['CreationTimestamp'] >= cutoff_date]
df_filtered
```

```python
df_sorted = df_filtered.sort_values(by='CreationTimestamp')
df_sorted
```

```python
coil.isnull().sum()
```

```python
coil = coil.fillna(value={'CoilType': 'Not Available','isReplacementFor': 0, 'LeadLocation': 'Unknown',
                         'LeadLocationLPOS': 'Not Specified', 'LeadType': 'Unknown', 'CoilAttributeID': 'Not Specified',
                         'ProductNumber': 'Not Specified', 'Status': 'Not Specified', 'ThrowDirection': 'Unknown'})
```

```python
coil.isnull().sum()
```

```python
coil[coil.Turns.isnull()]
```

```python
#outliers in Turns
plt.figure(figsize=(12, 6))
sns.boxplot(x='Turns', data=coil)
plt.title('Boxplot of Turns')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(coil['Turns'], bins=30, kde=True)
plt.title('Distribution of Turns')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='LeadType', hue='ThrowDirection', data=coil)
plt.title('LeadType vs. ThrowDirection')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='CoilType', hue='Status', data=coil)
plt.title('CoilType vs. Status')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='CoilType', data=coil)
plt.title('Distribution of CoilTypes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='LeadLocation', data=coil)
plt.title('Distribution of LeadLocations')
plt.xticks(rotation=45)
plt.show()
```

```python
coil.describe()
```

```python
coiltypes_counts=coil['CoilType'].value_counts()
coiltypes_counts
```

```python
coiltypes_counts = coil['CoilType'].value_counts()

coiltypes_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Coil Types')
plt.axis('equal') 
plt.show()
```
